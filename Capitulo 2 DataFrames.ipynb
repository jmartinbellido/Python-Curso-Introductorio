### Primeros pasos con un DataFrame

### Seleccionar filas seg칰n posiciones

Podemos utilizar una sintaxis similar a la que acabamos de introducir para filtrar filas seg칰n posiciones. 

```

```

Repetiremos esta misma estructura pero, esta vez, no utilizaremos input de tipo texto, sino un *list con booleanos y tantos elementos como filas en el DataFrame*.





````
# el siguiente df (DataFrame) contiene tres filas
# utilizamos un list con tres elementos booleanos
df_cars[[True, False, True]]
````

Las filas que se correspondan con valores *False* en nuestro list quedar치n omitidas de la selecci칩n. En otras palabras, conservaremos 칰nicamente las filas que se correspondan con valores *True* en nuestro list.

> 游녤 En este caso, tenemos un DataFrame con tan solo 3 filas. En la pr치ctica, operamos regularmente con DataFrames de miles o millones de filas; evidentemente no ser칤a pr치ctico tener que crear un list con tantos elementos booleanos para poder filtrar filas.

> 游녤 Retomaremos este tema cuando introduzcamos filtrar filas utilizando condiciones l칩gicas


### Seleccionar filas y columnas utilizando los m칠todos *.loc[ ]* y *.iloc[ ]*

A continuaci칩n, introducimos dos m칠todos de Pandas que buscan simplificar la tarea de filtrar filas y columnas. Los m칠todos *.loc[ ]* y *.iloc[ ]* permiten filtrar filas y columnas seg칰n etiquetas (nombres) y posiciones, respectivamente.

> 游녤 Estos m칠todos permiten filtrar filas y columnas siguiendo una l칩gica de matrices 

> 丘멆잺 Por excepci칩n, definimos estos dos m칠todos utilizando brackets [ ] y no par칠ntesis ( ). Esto es una excepci칩n que aplica 칰nicamente para estos dos m칠todos. Los creadores de Pandas han intentado

#### M칠todo loc[ ]

El method *.loc[]* permite acceder a filas y columnas utilizando nombres.

```
object.loc[rows,columns]
object.loc[[row_1,row_2,...],[col_1,col_2,...]]
```

El m칠todo contiene dos par치metros, el primero de ellos es donde definimos (filtramos) filas, el segundo donde filtramos columnas.

````
# visualizamos las primeras filas de nuestro df
df_jamesbond.head()
# podemos comprobar que tenemos un 칤ndice num칠rico (se genera de forma autom치tica)
````

````
# para este ejemplo, convertiremos la columna Film en nuestro 칤ndice
df_jamesbond = df_jamesbond.set_index("Film") # cambiamos 칤ndice, sobreescribimos objeto
df_jamesbond.head() # visualizamos primeras entradas
````

````
# a continuaci칩n, utilizaremos el m칠todo .loc[] para seleccionar elementos en nuestro df
df_jamesbond.loc[["From Russia with Love","Goldfinger"],:] # seleccionamos dos filas seg칰n nombre, todas las columnas
````

> 游녤 Hemos utilizado un list con dos elementos de tipo texto en el primer par치metro, donde filtramos filas

> 游녤 Hemos utilizado un *wildcard* (comod칤n) : (dos puntos) en el par치metro de las columnas para indicar que queremos conservar todas las columnas

````
# otro ejemplo
df_jamesbond.loc[:'Thunderball',['Actor','Director']]
````

> 游녤 Hemos utilizado el *wildcard* antes del valor *Thunderball* para indicar que queremos esta fila y todas las anteriores

> 游녤 Utilizamos un list en el segundo par치metro para filtrar dos columnas

#### M칠todo iloc[ ]
El m칠todo *.iloc[]* es completamente an치logo al anteerior, pero este permite filtrar filas y columnas seg칰n su posiciones.

```
object.iloc[row_position,column_position]
``` 

````
# volvemos a cargar DataFrame
df_jamesbond = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv")
df_jamesbond.head() # observamos que volvemos a tener 칤ndice num칠rico; tambi칠n recuperamos Film como columna
````

````
# a continuaci칩n, seleccionamos elementos seg칰n su posici칩n en el dataframe
df_jamesbond.iloc[:5,[1,3,5]] # seleccionamos todas las filas hasta la 5ta; columnas 1,3,5
````

> 游녤 Hemos seleccionados todas las filas hasta la 5ta; columnas 1,3,5

````
df_jamesbond.iloc[5:10,:]
````

> 游녤 En este segundo ejemplo, hemos filtrado filas seg칰n un rango (6ta a d칠cima) y conservado todas las columnas



En ocasiones, combinamos ambos m칠todos para, por ejemplo, filtrar elementos seg칰n posiciones, luego filtrar seg칰n etiquetas. Para esto utilizamos ambos m칠todos de forma anidada.

````
# anidamos ambos m칠todos 
df_jamesbond.iloc[:10,:].loc[:,['Film','Actor','Director']]
````

### Filtrar filas, seg칰n condici칩n l칩gica

Hasta ahora, hemos estado filtrando filas en nuestro DataFrame seg칰n sus 칤ndices y posiciones. Frecuentemente, nos interesa filtrar filas seg칰n condiciones l칩gicas. 

```
object[object[condition]]
```

````
# comenzaremos creando un objeto con elementos booleanos a partir de una prueba l칩gica
cond = df_jamesbond['Year'] >= 1990 # usamos operadores de comparaci칩n sobre una columna
cond
````

````
# utilizamos el objeto con booleanos como selecci칩n dentro de nuestro DataFrame, logrando filtrar filas
df_jamesbond[cond]
````

````
# lo anterior es equivalente a la siguiente operaci칩n
df_jamesbond[df_jamesbond['Year'] >= 1990]
# en este caso, no hemos necesitado ning칰n objeto auxiliar
````

````
# creamos condiciones y las almacenamos en objetos
cond_1 = df_jamesbond["Budget"]>150
cond_2 = df_jamesbond["Actor"]=='Daniel Craig'

# relacionamos las condiciones con un operador de tipo AND (&)
df_jamesbond[cond_1 & cond_2] # seleccionamos filas que cumplen con ambas condiciones de forma simult치nea
````

````
# creamos condiciones y las almacenamos en objetos
cond_1 = df_jamesbond["Budget"]>150
cond_2 = df_jamesbond["Actor"]=='Daniel Craig'

# relacionamos las condiciones con un operador de tipo OR (|)
df_jamesbond[cond_1 | cond_2] # seleccionamos filas que cumplan con al menos una condici칩n
````

````
# creamos condiciones y las almacenamos en objetos
cond_1 = df_jamesbond["Box Office"]>500
cond_2 = df_jamesbond["Actor"]=='Daniel Craig'

# relacionamos las condiciones con un operador de tipo AND (&)
# en esta ocasi칩n, negamos la condici칩n 2
df_jamesbond[cond_1 & -cond_2] 
````

#### M칠todos 칰tiles al crear condiciones para filtrar filas

````
# podemos utilizar el m칠todo .isin() para filtrar por m칰ltiples valores en una variable categ칩rica
cond = df_jamesbond['Director'].isin(['Martin Campbell','Terence Young']) # filtramos por m치s de un valor para una variable categ칩rica
df_jamesbond[cond]
````

````
# creamos condiciones para reconocer patrones espec칤ficos de texto
cond_1 = df_jamesbond['Film'].str.contains('love', regex=False, case=False)
cond_2 = df_jamesbond['Film'].str.contains('die', regex=False, case=False)

# creamos una condici칩n, reconociendo un patr칩n en un campo de tipo texto
df_jamesbond[cond_1 | cond_2]
````

### Crear una nueva columna a partir de una prueba l칩gica

A continuaci칩n, crearemos una nueva columna a partir de una condici칩n *if/else* utilizando la funci칩n *np.where()*. En este caso, el valor del campo dependar치 de una *condici칩n l칩gica*, no de una operaci칩n matem치tica.

### Python method chaining

Al anidar m칰ltiples operaciones, con frecuencia podemos encontrarnos realizando l칤neas de c칩digo demasiado extensas. Ante estas situaciones, podemos optar por separar nuestras operaciones en nuevas l칤neas de c칩digo utilizando una sintaxis especial llamada *method chaining*.

```
df.iloc[x,y]\
  .head()\
  .sort_values(x)
```
* utilizamos una barra invertida (\) para separar operaciones
* la barra invertida se coloca tras terminar una operaci칩n
* tras la barra invertida, damos enter para cambiar de l칤nea de c칩digo; no debemos dejar ning칰n caracter adicional

> 丘멆잺 *Method chaining no admite ning칰n caracter tras la barra invertida, tampoco comentarios*

````
# comenzamos realizando operaciones anidadas sin utilizar el method chanining
pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv").iloc[0:5,:].head().sort_values("Film")
````

````
# repetimos las cuatro operaciones anteriores, esta vez utilizando el method chaining
# este m칠todo adquiere mayor utilidad cuanto m치s operaciones realicemos
pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv")\
  .iloc[0:5,:]\
  .head()\
  .sort_values("Film")
````
