### Ejercicio #1
Debajo hay un list que contiene nombres de comida. Crear un *for loop* donde se despliegue "I love ... (cada elemento en el list)"

Resultado esperado:  
I love pizza  
I love sushi  
I love paella  
I love tapas  

>  Puedes concatenar texto facilmente uniendo inputs de tipo texto con un operador +




```
# ejemplo de concatenar texto
my_name = 'Martin'
print('Mi nombre es ' + my_name)
```

```
# utiliza el siguiente list para realizar el ejercicio
food_list = ['pizza','sushi','paella','tapas']
```

```
# resoluci贸n utilizando for loop
for f in food_list:
  print('I love ' + f)
```

```
# resoluci贸n utilizando comprehension lists
[print('I love ' + f) for f in food_list]
```

### Ejercicio #2
A continuaci贸n, presentamos otro list que contiene comida. Repetir el ejercicio anterior, esta vez evitar desplegar los elementos "onions" y "garlic" (sin editar el list). 

```
food_list = ['pizza','sushi','onion','paella','garlic','tapas']
```

```
# resoluci贸n utilizando operador IN
for f in food_list:
  if f not in ['onion','garlic']:
    print('I love ' + f)
```

```
# resoluci贸n utilizando operador l贸gico AND para combinar dos condiciones
for f in food_list:
  if not f == 'onion' and not f == 'garlic':
    print('I love ' + f)
```
