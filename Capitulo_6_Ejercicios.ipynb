### Ejercicio #1

Construir un gráfico de dispersión utilizando Matplotlib para visualizar *IMDB Score* vs *Budget* (ver imagen).


* Incluir solo películas propias de los 3 directores con mayor cantidad de películas  
* Utilizar colores y marcadores para diferenciar entre directores

> https://data-wizards.s3.amazonaws.com/datasets/movies.csv

![title](https://data-wizards.s3.amazonaws.com/img/matplotlib/EX1.png)


```python
# importamos librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importamos dataframe
df_movies = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/movies.csv")
df_movies.dtypes
```

```python
# obtenemos frecuencia para directores
freq_directors = df_movies['director_name'].value_counts()
select_directors = freq_directors.reset_index()['index']
```

```python
# modificamos variable para expresarla en MM
df_movies['budget'] = df_movies['budget']/1000000
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

# director 1
ax.scatter(
    x = df_movies[df_movies['director_name']==select_directors[0]]['budget'], 
    y = df_movies[df_movies['director_name']==select_directors[0]]['imdb_score'], 
    s=50, 
    facecolor='blue', 
    edgecolor='blue',
    marker='o',
    label=select_directors[0]
)

# director 2
ax.scatter(
    x = df_movies[df_movies['director_name']==select_directors[1]]['budget'], 
    y = df_movies[df_movies['director_name']==select_directors[1]]['imdb_score'], 
    s=50, 
    facecolor='red', 
    edgecolor='red',
    marker='^',
    label=select_directors[1]
)

# director 3
ax.scatter(
    x = df_movies[df_movies['director_name']==select_directors[2]]['budget'], 
    y = df_movies[df_movies['director_name']==select_directors[2]]['imdb_score'], 
    s=50, 
    facecolor='green', 
    edgecolor='green',
    marker='v',
    label=select_directors[2]
)

ax.set(xlabel="Budget (USD Million)",
       ylabel="IMDB Score",
       title="Budget vs. IMDB Score\nTop 3 Directores with Highest Number of Movies")

ax.legend()
```

### Ejercicio #3

El dataset contiene información histórica para un ecommerce. Construir dos gráficos de serie temporal para visualizar transacciones y visitas semanales para los primeros 6 meses de 2017.

> https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv

![title](https://data-wizards.s3.amazonaws.com/img/matplotlib/EX3.png)

```python
# importamos librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importamos dataframe
df_google_analytics = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv")
df_google_analytics.dtypes
```

```python
# importamos dataframe
df_google_analytics['date'] = pd.to_datetime(df_google_analytics['date'])
df_google_analytics.dtypes
```

```python
# creamos nuevos campos
df_google_analytics['week'] = df_google_analytics['date'].dt.isocalendar().week
df_google_analytics['month'] = df_google_analytics['date'].dt.month
df_google_analytics['year'] = df_google_analytics['date'].dt.year
df_google_analytics.head()
```

```python
# filtramos y manipulamos df
cond_1 = df_google_analytics.year == 2017
cond_2 = df_google_analytics.month < 7
cond_3 = df_google_analytics.week==52

df = df_google_analytics[cond_1 & cond_2 & -cond_3]\
  .groupby('week')\
  .agg({'transactions':'sum','visits':'sum'})\
  .reset_index()

df.head()
```

```python
# gráficos vacío
fig, ax = plt.subplots(2,1,figsize=(10,8))

# gráfico 1
ax[0].plot(
    df['week'],
    df['transactions'],
    color = 'red',
    linewidth=1.5
)

# otras configuraciones
ax[0].set(
    title = 'H1 2017\nWeekly Transactions',
    ylabel = 'Transactions',
    xticks = np.arange(0,26,step=1)
)

# gráfico 2
ax[1].plot(
    df['week'],
    df['visits']/1000,
    color = 'blue',
    linewidth=1.5
)

# otras configuraciones
ax[1].set(
    title = 'Weekly Visits',
    ylabel = 'Visits (K)',
    xlabel = 'Week Number',
    xticks = np.arange(0,26,step=1)
)
```
