### Primeros pasos

```python
# creamos un documento pdf vac칤o en memoria
my_pdf = FPDF()

# creamos una primer p치gina en blanco
my_pdf.add_page()

# creamos un t칤tulo
my_pdf.set_font(family='Arial', style='B', size=16) # we specify font
my_pdf.cell(w=0, h=10, txt='My First PDF', border = 1, align='C') # we include our first cell/container

# exportamos documento
my_pdf.output('pdf_dummy_1.pdf')
```

### Documento con gr치fico
